## Group By

7. ¿Cuántos préstamos hay por cada estado y cual es la media de la cantidad de dinero prestado?

```python
%%sql
select status, count(*) total, round(avg(amount), 2) as "Media prestado"
from LOAN
group by status -- comando para agrupar terminos
order by total desc
```

```python
%%sql
select status, count(*) total, round(avg(amount), 2) as "Media prestado"
from LOAN
where duration >24
group by status
order by total desc
```

```python
%%sql
select status, count(*) total, round(avg(amount), 2) as "Media prestado"
from LOAN
where duration >24
group by status
having total >25
order by total desc
```

El where se utiliza para filtrar antes de agrupaciones y el having despues de agrupaciones

10. ¿Cuántas cuentas de frecuencia mensual tienen un un préstamo activo?



hacerlo por inner join o hacerlo por where es lo mismo

si se mete un + al lado del igual seria como un left o un right join 

```python
#Query par saber las cuentas con frecuencia mensual y diversos datos
%%sql
select loan_id, a.date loan_date, amount, duration, payments, frequency, b.date acc_date
from loan a, account b 
where a.account_id = b.account_id
and status in ("A", "B")
and frequency is "monthly" -- frecuencia 
order by amount desc, a.date desc
limit 12
```

(limitar a 10)

- ¿Cuántos clientes hay en Praga?
- ¿Cuántas cuentas tienen asociados clientes en Praga?
- ¿cuántas de esas cuentas tienen asociados un préstamo?
- Necesito saber cuántos hay por cada estado (*status*)


```python
#a. ¿Cuántos clientes hay en Praga?

%%sql

select count(*)
from  DISTRICT di, CLIENT cl
where di.A1 = cl.district_id
and di.A3 = "Prague"
```

```python
#b. ¿Cuántas cuentas tienen asociados clientes en Praga? -> entiendo que cuantas cuentas hay en total en praga

%%sql

select count(*)
from  DISTRICT di, ACCOUNT ac
where di.A1 = ac.district_id
and di.A3 = "Prague"
```

```python
#c. ¿cuántas de esas cuentas tienen asociados un préstamo?

%%sql

select count(*)
from  DISTRICT di, ACCOUNT ac, loan lo
where di.A1 = ac.district_id
and ac.account_id = lo.account_id
and di.A3 = "Prague"
```

```python
#d. Necesito saber cuántos hay por cada estado (status)

%%sql

select lo.status, count(status)
from  DISTRICT di, ACCOUNT ac, loan lo
where di.A1 = ac.district_id
and ac.account_id = lo.account_id
and di.A3 = "Prague"
group by lo.status
```
